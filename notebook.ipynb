```python
# Intelligent Job Recommendation System (Basic Version)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# -------------------------------
# Step 1: User Resume
# -------------------------------
resume = "Python machine learning data analysis SQL"

# -------------------------------
# Step 2: Job Titles
# -------------------------------
job_titles = [
    "Data Scientist",
    "Web Developer",
    "Machine Learning Engineer",
    "Backend Developer"
]

# -------------------------------
# Step 3: Job Descriptions
# -------------------------------
jobs = [
    "Python machine learning data analysis",
    "HTML CSS JavaScript web development",
    "Python deep learning AI machine learning",
    "Java SQL backend development databases"
]

# -------------------------------
# Step 4: Combine Resume + Jobs
# -------------------------------
documents = [resume] + jobs

# -------------------------------
# Step 5: Apply TF-IDF
# -------------------------------
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

# -------------------------------
# Step 6: Compute Similarity
# -------------------------------
similarity = cosine_similarity(X[0:1], X[1:])

# -------------------------------
# Step 7: Sort and Recommend Jobs
# -------------------------------
scores = similarity[0]
sorted_indices = np.argsort(scores)[::-1]

print("Top Job Recommendations:\n")

for i in sorted_indices:
    print(f"{job_titles[i]} → Match Score: {scores[i]:.2f}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# -------------------------------
# Step 1: Skills list
# -------------------------------
skills_list = [
    "python", "machine learning", "data analysis", "sql",
    "java", "html", "css", "javascript", "deep learning", "ai"
]

# -------------------------------
# Step 2: User input
# -------------------------------
resume = input("Enter your resume/skills: ")

resume_lower = resume.lower()

# -------------------------------
# Step 3: Extract skills
# -------------------------------
extracted_skills = []

for skill in skills_list:
    if skill in resume_lower:
        extracted_skills.append(skill)

print("\nExtracted Skills:", extracted_skills)

# Convert to text
resume_skills_text = " ".join(extracted_skills)

# -------------------------------
# Step 4: Job Data
# -------------------------------
job_titles = [
    "Data Scientist",
    "Web Developer",
    "Machine Learning Engineer",
    "Backend Developer"
]

job_skills = [
    "python machine learning data analysis",
    "html css javascript",
    "python deep learning ai machine learning",
    "java sql backend"
]

# -------------------------------
# Step 5: TF-IDF
# -------------------------------
documents = [resume_skills_text] + job_skills

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

# -------------------------------
# Step 6: Similarity
# -------------------------------
similarity = cosine_similarity(X[0:1], X[1:])

# -------------------------------
# Step 7: Sort Results
# -------------------------------
scores = similarity[0]
sorted_indices = np.argsort(scores)[::-1]

# -------------------------------
# Step 8: Clean Output
# -------------------------------
print("\nTop Job Recommendations:\n")

for rank, i in enumerate(sorted_indices):
    print(f"{rank+1}. {job_titles[i]} → Match Score: {scores[i]:.2f}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# -------------------------------
# Step 1: Resume (fixed input)
# -------------------------------
resume = "Python machine learning data analysis SQL deep learning"

# -------------------------------
# Step 2: Extract important words (TF-IDF)
# -------------------------------
vectorizer1 = TfidfVectorizer()
X1 = vectorizer1.fit_transform([resume])

feature_names = vectorizer1.get_feature_names_out()
scores = X1.toarray()[0]

# Combine words with scores
word_scores = list(zip(feature_names, scores))

# Sort words by importance
word_scores = sorted(word_scores, key=lambda x: x[1], reverse=True)

# Take top 5 important words
top_skills = [word for word, score in word_scores[:5]]

print("Extracted Skills:", top_skills)

# Convert skills into text
resume_skills_text = " ".join(top_skills)

# -------------------------------
# Step 3: Job data
# -------------------------------
job_titles = [
    "Data Scientist",
    "Web Developer",
    "Machine Learning Engineer",
    "Backend Developer"
]

job_skills = [
    "python machine learning data analysis",
    "html css javascript",
    "python deep learning ai machine learning",
    "java sql backend"
]

# -------------------------------
# Step 4: TF-IDF for matching
# -------------------------------
documents = [resume_skills_text] + job_skills

vectorizer2 = TfidfVectorizer()
X2 = vectorizer2.fit_transform(documents)

# -------------------------------
# Step 5: Compute similarity
# -------------------------------
similarity = cosine_similarity(X2[0:1], X2[1:])

# -------------------------------
# Step 6: Sort results
# -------------------------------
scores = similarity[0]
sorted_indices = np.argsort(scores)[::-1]

# -------------------------------
# Step 7: Final output
# -------------------------------
print("\nTop Job Recommendations:\n")

for rank, i in enumerate(sorted_indices):
    print(f"{rank+1}. {job_titles[i]} → Match Score: {scores[i]:.2f}")
```
